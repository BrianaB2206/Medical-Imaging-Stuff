## Ex. 1

```python
x=np.array([[1,2,3,4,5,6],[11,12,13,14,15,16],[21,22,23,24,25,26]])
y = np.zeros((3,6))
for i in range(0,3):
    for j in range(0,6):
        if x[i,j]%2!=0:
            y[i,j]=x[i,j]
        else:
            y[i,j]=0
print('Y= ', y)
a=x.shape
b=y.shape
print('\nDim_X= ',a)
print('Dim_Y= ',b)
```

```python
from PIL import Image
img1 = Image.open('CThead.tif', 'r')
print('the size of the variable is: ', img1.size, 'and of type ', type(img1))
```

```python
img2 = np.asarray(img1)
img3 = img1.rotate(-70, translate = None)

matplotlib.pyplot.figure()
matplotlib.pyplot.subplot(1,2,1); matplotlib.pyplot.imshow(img2, cmap = 'gray'); matplotlib.pyplot.title('original')
matplotlib.pyplot.subplot(1,2,2); matplotlib.pyplot.imshow(np.asarray(img3), cmap = 'gray'); matplotlib.pyplot.title('rotit cu 70 de grade la dreapta')
```

```python
img4 = img1.rotate(0,translate=(40,0))
matplotlib.pyplot.figure()
matplotlib.pyplot.subplot(1,2,1); matplotlib.pyplot.imshow(img2, cmap = 'gray'); matplotlib.pyplot.title('original')
matplotlib.pyplot.subplot(1,2,2); matplotlib.pyplot.imshow(np.asarray(img4), cmap = 'gray'); matplotlib.pyplot.title('translatat cu 40 pixeli la dreapta')
```

```python
img5 = img1.rotate(0,translate=(0,-72))
matplotlib.pyplot.figure()
matplotlib.pyplot.subplot(1,2,1); matplotlib.pyplot.imshow(img2, cmap = 'gray'); matplotlib.pyplot.title('original')
matplotlib.pyplot.subplot(1,2,2); matplotlib.pyplot.imshow(np.asarray(img5), cmap = 'gray'); matplotlib.pyplot.title('translatat cu 72 pixeli in sus')
```
